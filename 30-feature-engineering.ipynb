## dataset: sessions_train, for simplicity, only keep locale = 'UK' and items appears more than 20 times

```python
sessions_train = pd.read_csv('/Users/stoneman/Library/CloudStorage/OneDrive-Vanderbilt/Social_Network_Analysis/Amazon_Recommandation_Project/data/sessions_train.csv')
```

```python
sessions_train
```

```python
sessions_train_UK = sessions_train[sessions_train['locale'] == 'UK']
```

```python
sessions_train_UK.shape
```

```python
from collections import Counter
import re

# First, let's concatenate prev_items and next_item into a single list
all_items = []
for index, row in sessions_train_UK.iterrows():
    # Use regular expression to match all alphanumeric sequences
    prev_items_list = re.findall(r"'([^']*)'", row['prev_items'])

    # Combine with the next_item
    combined_items = prev_items_list + [row['next_item']]
    all_items.extend(combined_items)

# Count the frequency of each item ID
item_counter = Counter(all_items)
```

```python
# Extract items that appear several times
# HERE I SET THE THRESHOLD TO 20
items_with_multiple_occurrences = [item for item, count in item_counter.items() if count > 100]
```

```python
len(items_with_multiple_occurrences)
```

```python
# Now, filter the DataFrame to only keep rows where 'next_item' is in 'items_with_multiple_occurrences'
sessions_train_UK_filtered = sessions_train_UK[sessions_train_UK['next_item'].isin(items_with_multiple_occurrences)]
sessions_train_UK_filtered.shape
```

```python
sessions_train_UK_filtered.drop('locale', axis=1, inplace=True)
```

```python
sessions_train_UK_filtered
```

```python
sessions_train_UK_filtered.to_csv('/Users/stoneman/Library/CloudStorage/OneDrive-Vanderbilt/Social_Network_Analysis/Amazon_Recommandation_Project/data/sessions_train_UK_filtered.csv', index=False)
```

```python
# Initialize an empty list to store the filtered prev_items
filtered_prev_items = []

# Loop over each row in the DataFrame
for index, row in sessions_train_UK_filtered.iterrows():
    # Use regular expression to match all alphanumeric sequences
    prev_items_list = re.findall(r"'([^']*)'", row['prev_items'])

    # Remove items not in items_with_multiple_occurrences
    filtered_list = [item for item in prev_items_list if item in items_with_multiple_occurrences]
    
    # Append the filtered list to the accumulator list
    filtered_prev_items.append(filtered_list)
```

```python
# Assign the filtered lists to a new column in the DataFrame
sessions_train_UK_filtered['pre_items_filter'] = filtered_prev_items
```

```python
sessions_train_UK_filtered.drop('prev_items', axis=1, inplace=True)
sessions_train_UK_filtered.rename(columns={'pre_items_filter': 'prev_items'}, inplace=True)
```

```python
df_UK_finalized = pd.read_csv('/Users/stoneman/Library/CloudStorage/OneDrive-Vanderbilt/Social_Network_Analysis/Amazon_Recommandation_Project/data/products_train_UK_finalized.csv')
```

```python
df_UK_finalized
```

```python
# Set the device to CPU
device = torch.device('cpu')

# Load the tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased').to(device)
model.eval()  # Put the model in evaluation mode

# Initialize the list to hold representations
representations = []

# Determine the number of chunks to process
total = len(df_UK_finalized)
chunk_size = 100  # Set this to something that fits in your memory
num_chunks = total // chunk_size + (1 if total % chunk_size != 0 else 0)

# Process the dataframe in chunks
for i in range(num_chunks):
    start_index = i * chunk_size
    end_index = start_index + chunk_size
    
    # Perform tokenization for the current chunk
    inputs = tokenizer(df_UK_finalized['description'].iloc[start_index:end_index].tolist(),
                       padding=True,
                       truncation=True,
                       max_length=512,
                       return_tensors="pt").to(device)
    
    # Forward pass to get embeddings
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Extract the pooled output
    paragraph_representations = outputs.pooler_output.detach().cpu().numpy().tolist()
    representations.extend(paragraph_representations)
    
    # Display the progress
    print(f"Chunk {i+1}/{num_chunks} processed. ({(i+1)/num_chunks:.2%} complete)")

# After processing all chunks, add the representations to the dataframe
df_UK_finalized['representation'] = representations
```
